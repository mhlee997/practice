```python
# 2번
table = ["XOXO", "OXXO", "XXOX", "XOOO"]
table_data = []
for t in table:
    r = 0
    for c in range(len(t)):
        r = r << 1
        if t[c] == 'O':
            r += 1
    table_data.append(r)

product_num = int(len(table))

ans = 0
array = list(range(len(table)))
ret = []

def ok():
    global ret, table_data
    value = 0
    for r in ret:
        value = value | table_data[r]
    if value == ((1 << product_num) - 1):
        return True
    return False

def combination(n, ret_idx, array_idx):
    global ret, array, ans
    if ans:
        return
    if ret_idx == n:
        if ok():
            ans = n
        return
    if array_idx > len(array):
        return
    for idx in range(array_idx, len(array)):
        ret.append(array[idx])
        combination(n, ret_idx+1, idx+1)
        ret.pop()

for i in range(1, product_num + 1):
    ret = []
    combination(i, 0, 0)
    if ans:
        break

print(ans) 
```

```python
# 3번
costs = [10, 10, 5, 6]
edges = [[3, 1, 2], [1, 2, 1], [4, 1, 2], [4, 2, 3]]
n = len(costs)
edge_list = [[] for i in range(n + 1)]

par = [i for i in range(n + 1)]

def union(x, y):
    global par
    x = find(x)
    y = find(y)
    if x != y:
        par[y] = x

def find(x):
    global par
    if (x == par[x]):
        return x
    par[x] = find(par[x])
    return par[x]

for edge in edges:
    s = edge[0]
    e = edge[1]
    c = edge[2]
    par[s] = e
    edge_list[e].append([s, c])

cnt = [0 for i in range(n + 1)]
root = [False for i in range(n + 1)]
root_node = find(1)
cnt[root_node] = 1

for i in range(1, n + 1):
    if len(edge_list[i]) == 0:
        root[i] = True

def solve(node):
    num = len(edge_list[node])
    for idx in range(num):
        next = edge_list[node][idx][0]
        cost = edge_list[node][idx][1]
        cnt[next] += cost * cnt[node]
        solve(next)

solve(root_node)

mod = 10**9 + 7
material_cost = 0
personnel_expense = 0
for idx in range(1, n + 1):
    cost = (costs[idx-1] * cnt[idx]) % mod
    if root[idx]:
        material_cost = (material_cost + cost) % mod
    else:
        personnel_expense = (personnel_expense + cost) % mod

print(material_cost, personnel_expense)
```
